###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
for i in df['Department'].unique():
  x=df.loc[df['Department'] ==i]
  x.sort_values(by="CodeKata Score",ascending=False)
  a=x.head(5)
  a.plot.bar(x="Name", y="CodeKata Score", rot=90)
  print(i)
  print("\n")
  plt.show()
  print("\n")
```
